##### ü§ñReplyü§ñ<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->

Exactly right! You've got the concept.

**WireGuard specifics:**

- **Protocol:** UDP (User Datagram Protocol)
- **Port:** You choose it (default is 51820, but you can pick any available port)
- **How it works:** 
  - Your device wraps your regular traffic (IP packets) inside encrypted WireGuard packets
  - These encrypted packets get sent via UDP to your server's port 51820
  - The server unwraps (decrypts) them, extracts the original packets, and forwards them to the internet
  - Responses come back the same way in reverse

It's called "encapsulation" - your real traffic is the payload inside the encrypted WireGuard packet.

**Why UDP?** It's faster than TCP because it doesn't wait for acknowledgments. WireGuard is designed to be very lightweight and fast.

One practical note: If you're behind a very restrictive firewall that blocks unusual ports, you might want to use port 443 (HTTPS port) instead of 51820, since that's rarely blocked.
